```python
import pandas as pd
```

```python
training = pd.read_csv('2018-19.csv')
test = pd.read_csv('2020-2021.csv')
```

```python
training.columns
```

```python
def preprocessing(df):
  df.drop(['Div','Date','FTHG','FTAG','HTHG','HTAG','HTR','Referee', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC',
       'AC', 'HY', 'AY', 'HR', 'AR', 'B365H', 'B365D', 'B365A', 'BWH', 'BWD',
       'BWA', 'IWH', 'IWD', 'IWA', 'PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA',
       'VCH', 'VCD', 'VCA', 'Bb1X2', 'BbMxH', 'BbAvH', 'BbMxD', 'BbAvD',
       'BbMxA', 'BbAvA', 'BbOU', 'BbMx>2.5', 'BbAv>2.5', 'BbMx<2.5',
       'BbAv<2.5', 'BbAH', 'BbAHh', 'BbMxAHH', 'BbAvAHH', 'BbMxAHA', 'BbAvAHA',
       'PSCH', 'PSCD', 'PSCA'],axis =1,inplace = True)
  from sklearn.preprocessing import LabelEncoder
  lb = LabelEncoder()
  df['HomeTeam'] = lb.fit_transform(df['HomeTeam'])
  df['AwayTeam']=lb.fit_transform(df['AwayTeam'])
```

```python
test = test.reindex(columns=training.columns,fill_value=0)
```

```python
preprocessing(training)
preprocessing(test)
```

```python
training.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
X_train = training.drop(['FTR'],axis=1)
y_train = training['FTR']

X_test = test.drop(['FTR'],axis =1)
y_test = test['FTR']
```

```python
model.fit(X_train,y_train)
predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,predict)
accuracy
```
